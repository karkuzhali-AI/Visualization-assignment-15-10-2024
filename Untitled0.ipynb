```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate a synthetic dataset with correlated features
np.random.seed(42)
n = 500  # Number of samples

# Generate some correlated features
X1 = np.random.randn(n)  # Random values from a normal distribution
X2 = 2 * X1 + np.random.randn(n) * 0.5  # Strong positive correlation with X1
X3 = -3 * X1 + np.random.randn(n) * 0.2  # Strong negative correlation with X1
X4 = np.random.randn(n)  # No strong correlation with X1

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': X1,
    'Feature 2': X2,
    'Feature 3': X3,
    'Feature 4': X4
})

# Step 2: Compute the correlation matrix
corr_matrix = data.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the sales dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Pivot the data to create a grid for the heatmap
heatmap_data = pd.pivot_table(df, values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', linewidths=0.5)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Day')
plt.title('Heatmap of Sales Variation Across Months and Days')

# Show the plot
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z based on the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame (optional for further analysis)
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate the dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

# Create the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'],
            s=df['Population'] * 5,  # Bubble size proportional to population
            alpha=0.5,
            c='blue',
            edgecolors='w',
            linewidth=2)

# Add labels and title
for i in range(len(df)):
    plt.annotate(df['Country'][i],
                 (df['GDP'][i], df['Population'][i]),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')

plt.title('Bubble Chart of Country Population vs GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')

# Show the plot
plt.grid()
plt.show()

```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values between 0 and 100
y = np.random.rand(n) * 100  # Random y values between 0 and 100
sizes = np.random.randint(10, 100, n)  # Random sizes for the markers
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], n)  # Random colors

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh scatter plot
p = figure(title="Bokeh Scatter Plot", x_axis_label='X Values', y_axis_label='Y Values', width=800, height=400)

# Add scatter points to the plot
p.scatter(x='x', y='y', size='sizes', color='colors', source=data, legend_field='colors', alpha=0.6)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh histogram
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency',
           width=800, height=400)

# Add the histogram bars to the plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color='blue', line_color='black', alpha=0.7)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.palettes import Viridis256

# Prepare the output for Jupyter Notebook
output_notebook()

# Create a random dataset for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a meshgrid for the heatmap
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for plotting
data_flat = data_heatmap.flatten()
x_flat = xx.flatten()
y_flat = yy.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x_flat, 'y': y_flat, 'value': data_flat})

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           width=400, height=400, tools="")

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=data_flat.min(), high=data_flat.max())

# Create a heatmap with squares
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        line_color=None, fill_color=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
